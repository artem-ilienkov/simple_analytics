```python
# 2. Загрузим необходимые таблицы, преобразуем данные
accruals = gc.open('3_accruals').sheet1
cost_price = gc.open('2_cost_price').sheet1
adv_traffic = gc.open('3_adv_traffic').sheet1
adv_search = gc.open('4_adv_search').sheet1
review_points = gc.open('5_review_points').sheet1
review_fixing = gc.open('6_review_fixing').sheet1
sku = gc.open('sku').sheet1

accruals = accruals.get_all_values()
cost_price = cost_price.get_all_values()
adv_traffic = adv_traffic.get_all_values()
adv_search = adv_search.get_all_values()
review_points = review_points.get_all_values()
review_fixing = review_fixing.get_all_values()
sku = sku.get_all_values()


accruals = pd.DataFrame.from_records(accruals[2:],columns=accruals[1])
cost_price = pd.DataFrame.from_records(cost_price[2:],columns=cost_price[1])
adv_traffic = pd.DataFrame.from_records(adv_traffic[2:],columns=adv_traffic[1])
adv_search = pd.DataFrame.from_records(adv_search[2:],columns=adv_search[1])
review_points = pd.DataFrame.from_records(review_points[2:],columns=review_points[1])
review_fixing = pd.DataFrame.from_records(review_fixing[2:],columns=review_fixing[1])
sku = pd.DataFrame.from_records(sku[2:],columns=sku[1])


print('Таблицы для анализа сформированы')

# 2.1 Форматирование таблицы Accurals
# accruals['date_accrual'] = pd.to_datetime(accruals.date_accrual,infer_datetime_format=True)
accruals['date_accrual'] = pd.to_datetime(accruals.date_accrual,dayfirst=True)
accruals['date_order'] = pd.to_datetime(accruals.date_order,dayfirst=True)
accruals['quantity_item'] = accruals['quantity_item'].astype('int')
accruals['final_price'] = accruals['final_price'].str.replace(',', '.')
accruals['commission_sale'] = accruals['commission_sale'].str.replace(',', '.')
accruals['order_assembly'] = accruals['order_assembly'].str.replace(',', '.')
accruals['shipment_processing'] = accruals['shipment_processing'].str.replace(',', '.')
accruals['highway'] = accruals['highway'].str.replace(',', '.')
accruals['last_mile'] = accruals['last_mile'].str.replace(',', '.')
accruals['return_highway'] = accruals['return_highway'].str.replace(',', '.')
accruals['refund_processing'] = accruals['refund_processing'].str.replace(',', '.')
accruals['canceled_processing'] = accruals['canceled_processing'].str.replace(',', '.')
accruals['non_purchase_processing'] = accruals['non_purchase_processing'].str.replace(',', '.')
accruals['logistics'] = accruals['logistics'].str.replace(',', '.')
accruals['localization_index'] = accruals['localization_index'].str.replace(',', '.')
accruals['reverse_logistics'] = accruals['reverse_logistics'].str.replace(',', '.')
accruals['revenue'] = accruals['revenue'].str.replace(',', '.')
accruals['final_price'] = accruals['final_price'].astype('float64')
accruals['commission_sale'] = accruals['commission_sale'].astype('float64')
accruals['order_assembly'] = accruals['order_assembly'].astype('float64')
accruals['shipment_processing'] = accruals['shipment_processing'].astype('float64')
accruals['highway'] = accruals['highway'].astype('float64')
accruals['last_mile'] = accruals['last_mile'].astype('float64')
accruals['return_highway'] = accruals['return_highway'].astype('float64')
accruals['refund_processing'] = accruals['refund_processing'].astype('float64')
accruals['canceled_processing'] = accruals['canceled_processing'].astype('float64')
accruals['non_purchase_processing'] = accruals['non_purchase_processing'].astype('float64')
accruals['logistics'] = accruals['logistics'].astype('float64')
accruals['reverse_logistics'] = accruals['reverse_logistics'].astype('float64')
accruals['revenue'] = accruals['revenue'].astype('float64')
# accruals = accruals[['date_accrual', 'type_accrual', 'order_id', 'date_order','sku','article_id','name_product','quantity_item','final_price','profit']]

# 2.2 Форматирование таблицы cost_price
cost_price['start_date'] = pd.to_datetime(cost_price.start_date,dayfirst=True)
cost_price['end_date'] = pd.to_datetime(cost_price.end_date,dayfirst=True)
cost_price['price'] = cost_price['price'].str.replace(',', '.')
cost_price['delivery_fpp'] = cost_price['delivery_fpp'].str.replace(',', '.')
cost_price['cost_price'] = cost_price['cost_price'].str.replace(',', '.')
cost_price['price'] = cost_price['price'].astype('float64')
cost_price['delivery_fpp'] = cost_price['delivery_fpp'].astype('float64')
cost_price['cost_price'] = cost_price['cost_price'].astype('float64')

# 2.3 Форматирование таблицы adv_traffic
adv_traffic['date_order'] = pd.to_datetime(adv_traffic.date_order,dayfirst=True)
adv_traffic['final_price_adv'] = adv_traffic['final_price_adv'].str.replace(',', '.')
adv_traffic['impressions'] = adv_traffic['impressions'].str.replace(',', '.')
adv_traffic['clicks'] = adv_traffic['clicks'].str.replace(',', '.')
adv_traffic['ctr'] = adv_traffic['ctr'].str.replace(',', '.')
adv_traffic['avg_cpc'] = adv_traffic['avg_cpc'].str.replace(',', '.')
adv_traffic['expenses_target'] = adv_traffic['expenses_target'].str.replace(',', '.')
adv_traffic['quantity_orders'] = adv_traffic['quantity_orders'].str.replace(',', '.')
adv_traffic['profit_item'] = adv_traffic['profit_item'].str.replace(',', '.')
adv_traffic['quantity_item'] = adv_traffic['quantity_orders'].str.replace(',', '.')
adv_traffic['revenue_adv'] = adv_traffic['revenue_adv'].str.replace(',', '.')
adv_traffic['final_price_adv'] = adv_traffic['final_price_adv'].astype('float64')
adv_traffic['impressions'] = adv_traffic['impressions'].astype('float64')
adv_traffic['clicks'] = adv_traffic['clicks'].astype('float64')
adv_traffic['ctr'] = adv_traffic['ctr'].astype('float64')
adv_traffic['avg_cpc'] = adv_traffic['avg_cpc'].astype('float64')
adv_traffic['expenses_target'] = adv_traffic['expenses_target'].astype('float64')
adv_traffic['quantity_orders'] = adv_traffic['quantity_orders'].astype('float64')
adv_traffic['profit_item'] = adv_traffic['profit_item'].astype('float64')
adv_traffic['quantity_item'] = adv_traffic['quantity_item'].astype('float64')
adv_traffic['revenue_adv'] = adv_traffic['revenue_adv'].astype('float64')

# 2.4 Форматирование таблицы adv_search
adv_search['final_price'] = adv_search['final_price'].str.replace(',', '.')
adv_search['commission_rate'] = adv_search['commission_rate'].str.replace(',', '.')
adv_search['commission_rub'] = adv_search['commission_rub'].str.replace(',', '.')
adv_search['expenses_search'] = adv_search['expenses_search'].str.replace(',', '.')
adv_search['final_price'] = adv_search['final_price'].astype('float64')
adv_search['commission_rate'] = adv_search['commission_rate'].astype('float64')
adv_search['commission_rub'] = adv_search['commission_rub'].astype('float64')
adv_search['expenses_search'] = adv_search['expenses_search'].astype('float64')
adv_search['quantity_search'] = adv_search['quantity_search'].astype('float64')

# 2.5 Форматирование таблицы review_points
review_points['date_order'] = pd.to_datetime(review_points.date_order,dayfirst=True)
review_points['date_points'] = pd.to_datetime(review_points.date_points,dayfirst=True)
review_points['expenses_review'] = review_points['expenses_review'].str.replace(',', '.')
review_points['expenses_review'] = review_points['expenses_review'].astype('float64')

# 2.6 Форматирование таблицы review_fixing
review_fixing['date_order'] = pd.to_datetime(review_fixing.date_order) #,dayfirst=True)
review_fixing['expenses_fixing'] = review_fixing['expenses_fixing'].str.replace(',', '.')
review_fixing['expenses_fixing'] = review_fixing['expenses_fixing'].astype('float64')

print('Таблицы отформатированы')

# 3 Создадим таблицу по доставкам, без возвратов, отмен и иных статусов не связанных с доставкой

profit_margin = accruals.copy()
profit_margin = profit_margin[['date_accrual','type_accrual','order_id',
                               'date_order','sku','article_id',
                               'quantity_item','final_price','commission_sale','revenue',
                               'order_assembly', 'shipment_processing', 'highway',
                               'last_mile', 'return_highway', 'refund_processing',
                               'canceled_processing', 'non_purchase_processing', 'logistics',
                               'reverse_logistics']]

# Считаем "Обработка и доставка"
profit_margin['process_delivery'] = (profit_margin.order_assembly + profit_margin.shipment_processing \
                                    + profit_margin.highway + profit_margin.last_mile + profit_margin.canceled_processing \
                                    + profit_margin.non_purchase_processing + profit_margin.logistics)
# Считаем "Возвраты и отмены"
profit_margin['refund_cancel'] = (profit_margin.return_highway + profit_margin.refund_processing \
                                    + profit_margin.reverse_logistics)

profit_margin = profit_margin.drop(['order_assembly', 'shipment_processing', 'highway', 'last_mile', 'return_highway',
                                   'refund_processing', 'canceled_processing', 'non_purchase_processing', 'logistics', 'reverse_logistics'], axis=1)

profit_margin['order_id'] = profit_margin['order_id'].str.split(r'(\d{6,10}-\d\d\d\d)', expand=True)[1]

profit_margin = profit_margin[['date_accrual','type_accrual','order_id',
                               'date_order','sku','article_id',
                               'quantity_item','final_price','commission_sale',
                               'process_delivery', 'refund_cancel', 'revenue']]

# 3.1.1 Создание таблицы по статусам, где есть номер заказа
profit_margin_with_order_id = profit_margin.copy()
profit_margin_with_order_id = profit_margin_with_order_id.query('type_accrual != "Закрепление отзыва" \
                                     and type_accrual != "Услуги продвижения товаров" \
                                     and type_accrual != "Начисления по претензиям" \
                                     and type_accrual != "Доставка товаров на склад Ozon (кросс-докинг)" \
                                     and type_accrual != "Premium-подписка" \
                                     and type_accrual != "Приобретение отзывов на платформе" \
                                     and type_accrual != "Декомпенсации и возвращение товаров на сток" \
                                     and type_accrual != "Начисление за гибкий график выплат" \
                                     and type_accrual != "Утилизация" \
                                     and type_accrual != "Услуга досрочной выплаты" \
                                     and type_accrual != "Обработка товара в составе грузоместа на FBO" \
                                     and type_accrual != "Услуга по обработке опознанных излишков в составе ГМ" \
                                     and type_accrual != "Услуга по бронированию места и персонала для поставки с неполным составом в составе ГМ" \
                                     and type_accrual != "Услуга по обработке опознанных излишков" \
                                     and type_accrual != "Услуга по бронированию места и персонала для поставки с неполным составом" \
                                     and type_accrual != "Услуга за обработку операционных ошибок продавца: просроченная отгрузка" \
                                     and type_accrual != "Услуга за обработку операционных ошибок продавца: отмена"')

# 3.2 Создание таблицы с датами заказа
profit_date_order = profit_margin.copy()
profit_date_order = profit_date_order[['order_id','date_order']]
profit_date_order["rang"] = profit_date_order.groupby('order_id')['date_order'].cumcount()+1
profit_date_order = profit_date_order.query('rang == 1')
profit_date_order = profit_date_order.drop('rang', axis=1)

# 3.3 Соединение основной таблицы с таблицой с датами заказа. Получаем таблицу, которую можно разбивать на отдельные для анализа
profit_margin_with_order_id = profit_margin_with_order_id.merge(profit_date_order, how='left', on='order_id')
profit_margin_with_order_id = profit_margin_with_order_id.drop('date_order_x', axis=1)

# 3.4 Форматируем таблицу. Тут все заказы с доставкой и возвратом, за исключением типа начисления без номера заказа и без артикулов
profit_margin_with_order_id = profit_margin_with_order_id.rename(columns={'date_order_y': 'date_order'})

profit_margin_with_order_id = profit_margin_with_order_id[['date_order','date_accrual','type_accrual','order_id',
                               'sku','article_id', 'quantity_item','final_price','commission_sale',
                               'process_delivery', 'refund_cancel', 'revenue']]

# 3.5 Создание таблицы с количество каждого артикула в заказе
profit_margin_quantity = profit_margin.copy()
profit_margin_quantity = profit_margin_quantity.query('type_accrual == "Доставка покупателю"') \
                                               .groupby(['date_order', 'order_id','article_id']) \
                                               .agg({'quantity_item':'sum'}) \
                                               .rename(columns={'quantity_item': 'quantity'}).reset_index()
profit_margin_quantity = profit_margin_quantity[['order_id','article_id', 'quantity']]

# 4 Формирование агрегирующей таблицы по заказам
orders = profit_margin_with_order_id.copy()
orders = profit_margin_with_order_id.groupby(['date_order','order_id', 'article_id']).agg({'revenue':'sum'}).reset_index()

# 4.1 Присоединение к таблице данных quantity в заказе
orders = orders.merge(profit_margin_quantity, how='left', on=['order_id', 'article_id'])

# 4.2 Создание таблицы с final_price по дате заказа, номеру заказа и артикулу
profit_margin_final_price = profit_margin_with_order_id.copy()
profit_margin_final_price = profit_margin_final_price.groupby(['date_order','order_id', 'article_id']).agg({'final_price':'sum'}).reset_index()

# 4.3 Соединение таблицы orders с profit_margin_final_price
orders = orders.merge(profit_margin_final_price, how='left', on=['date_order','order_id', 'article_id'])

# 4.4 Создание таблицы с вознаграждением за продажу commission_sale
profit_margin_commission_sale = profit_margin_with_order_id.copy()
profit_margin_commission_sale = profit_margin_commission_sale.groupby(['date_order', 'order_id','article_id']).agg({'commission_sale':'sum'})

# 4.5 Соединение таблицы orders с profit_margin_logistics_costs
orders = orders.merge(profit_margin_commission_sale, how='left', on=['date_order','order_id', 'article_id'])

# 4.5.1 Создание таблицы с обработка и доставка process_delivery
profit_margin_process_delivery = profit_margin_with_order_id.copy()
profit_margin_process_delivery = profit_margin_process_delivery.groupby(['date_order', 'order_id','article_id']).agg({'process_delivery':'sum'})

# 4.5.2 Соединение таблицы orders с profit_margin_process_delivery
orders = orders.merge(profit_margin_process_delivery, how='left', on=['date_order','order_id', 'article_id'])

# 4.5.3 Создание таблицы с возвраты и отмены refund_cancel
profit_margin_refund_cancel = profit_margin_with_order_id.copy()
profit_margin_refund_cancel = profit_margin_refund_cancel.groupby(['date_order', 'order_id','article_id']).agg({'refund_cancel':'sum'})

# 4.5.4 Соединение таблицы orders с refund_cancel
orders = orders.merge(profit_margin_refund_cancel, how='left', on=['date_order','order_id', 'article_id'])

# 4.5.5 Заполним пропуски
orders = orders.fillna(0)

# 5 Добавление к таблице с заказами столбец с актуальной себестоимостью на дату заказа
# 5.1 Присоединяем себестоимость к заказам на основе article_id и даты заказа
merged_df = pd.merge(orders, cost_price, on='article_id', how='left')
# 5.2 Отфильтруем строки, чтобы заказ был сделан в период действия себестоимости
filtered_df = merged_df[(merged_df['date_order'] >= merged_df['start_date']) & (merged_df['date_order'] <= merged_df['end_date'])]
# 5.3 Уберем лишние столбцы
orders = filtered_df.drop(columns=['start_date', 'end_date', 'delivery_fpp', 'price'])

# 6 Создадим колонку - cost_price_2 в которой будет расчет себестоимости исходя их количества товаров в заказе и условий (где-то будет с минусом показатель, где-то ноль)
def f(row):
  if row['quantity'] > 0 and row['final_price'] > 0:
    return row['cost_price'] * row['quantity']
  elif row['quantity'] > 0 and row['final_price'] < 0:
    return (row['cost_price'] * -1) * row['quantity']
  elif row['final_price'] < 0:
    return row['cost_price'] * -1
  else:
    return 0
# 4.6.5 Применение функции к финальной таблице из п.3.9.3
orders['cost_price_2'] = orders.apply(f, axis=1)

# 7 Создадим колонку - gross_profit (валовая прибыль), в которой из колонки revenue (Итого) вычтем себестоимости с учетом кол-ва (cost_price_2)
def g(row):
  if row['revenue'] < 0 and row['final_price'] == 0:
    return row['revenue']
  elif row['revenue'] > 0 and row['final_price'] > 0:
    return row['revenue'] - row['cost_price_2']
  elif row['revenue'] < 0 and row['final_price'] < 0:
    return row['revenue'] + (row['cost_price_2'] * -1)
  else:
    return 0
# 4.6.5 Применение функции к финальной таблице из п.3.9.3
orders['gross_profit'] = orders.apply(g, axis=1)

# 8 Создадим таблицу, в которой будет кол-во позиций с разрезе дня и артикула - нужно для корректного распределения затрат на рекламу и отзывы
quantity_id = orders.copy()
quantity_id = quantity_id.groupby(['date_order', 'article_id']).agg({'order_id':'count'}).rename(columns = {'order_id':'quantity_id'}).reset_index()

# 8.1 Присоеденим к таблице с заказами данные с кол-вом артикулов на день из п.8
orders = orders.merge(quantity_id, how='left', on=['date_order','article_id'])

# 9 Сформируем таблицу с затратами на поисковую рекламу и присоеденим ее к таблице заказов (orders)
search = adv_search.copy()
search = search[['order_id','article_id','expenses_search']]
orders = orders.merge(search, how='outer', on=['order_id', 'article_id'])

# 10 Создадим колонку, в которой присвоим статусы заказам, по которым заказы были и была поисковая реклама, и заказов не было и была поисковая реклама
# 10.1 Создадим функцию, которая присовит статусы бонусам
def search_status(row):
  if pd.isna(row['date_order']):
    return 'search_without_sales'
  elif row['quantity'] >= 0 and pd.isna(row['expenses_search']):
    return 'sales_without_search'
  elif row['quantity'] >= 0 and row['expenses_search'] > 0:
    return 'sales_with_search'
  else:
    return 'empty'

orders['search_status'] = orders.apply(search_status, axis=1)

# 11 Присоеденим к таблице с заказами затраты на таргетированную рекламу

# 11.1 Привяжем к таблице с рекламными затратами артикулы
adv_traffic = adv_traffic.merge(sku, how='left', on='sku')
adv_traffic = adv_traffic[['date_order', 'article_id', 'expenses_target']]

# 11.2 Агрегируем затраты по-артикульно на каждый день
adv_traffic = adv_traffic.groupby(['date_order','article_id']).agg({'expenses_target':'sum'}).reset_index()

# 11.3 Присоеденим к таблице с заказами данные с рекламными затратами
orders = orders.merge(adv_traffic, how='outer', on=['date_order','article_id'])

# 11.4 Создадим колонку target_adv_status, в которой будут статусы - если были продажи и была реклама и была реклама, но не было продаж
def avg_status(row):
  if row['quantity'] >= 0 and pd.isna(row['expenses_target']):
    return 'sales_without_adv'
  elif row['quantity'] >= 0 and row['expenses_target'] == 0:
    return 'sales_without_adv'
  elif row['quantity'] >= 0 and row['expenses_target'] > 0:
    return 'sales_with_adv'
  elif pd.isna(row['order_id']):
    return 'adv_without_sales'
  else:
    return 'empty'

orders['target_status'] = orders.apply(avg_status, axis=1)

# 11.5 Создадим функцию, которыя рассчитает затраты на рекламу
def expenses_adv(row):
  if row['target_status'] == 'sales_without_adv':
    return row['expenses_target']
  elif row['target_status'] == 'sales_with_adv':
    return row['expenses_target'] / row['quantity_id']
  elif row['target_status'] == 'empty':
    return row['expenses_target']
  elif row['target_status'] == 'adv_without_sales':
    return row['expenses_target']
  else:
    return 'empty'
orders['expenses_target_2'] = orders.apply(expenses_adv, axis=1)

# 12 Присоеденим к таблице с заказами затраты на отзывы
# 12.1 Сформируем агрегирующую таблицу
review_points = review_points.groupby(['date_order','article_id']).agg({'expenses_review':'sum'}).reset_index()

# 12.2 Присоеденим к таблце данные с рекламными затратами
orders = orders.merge(review_points, how='outer', on=['date_order','article_id'])

# 12.3 Создадим функцию, которая присовит статусы бонусам
def review_status(row):
  if row['quantity'] == 0 and pd.isna(row['expenses_review']) and not pd.isna(row['order_id']):
    return 'sales_without_review'

  elif row['quantity'] > 0 and pd.isna(row['expenses_review']) and not pd.isna(row['order_id']):
    return 'sales_without_review'

  elif row['quantity'] == 0 and row['expenses_review'] > 0 and not pd.isna(row['order_id']):
    return 'sales_with_review'

  elif row['quantity'] > 0 and row['expenses_review'] > 0 and not pd.isna(row['order_id']):
    return 'sales_with_review'

  elif pd.isna(row['order_id']):
    return 'review_without_sales'

  elif pd.isna(row['date_order']):
    return 'empty'
  else:
    return 'error'

orders['review_status'] = orders.apply(review_status, axis=1)

# 12.4 Создадим функцию, которая распределит бонусы по заказам
def expenses_review(row):
  if row['review_status'] == 'sales_without_review':
    return row['expenses_review']
  elif row['review_status'] == 'review_without_sales':
    return row['expenses_review']
  elif row['review_status'] == 'empty':
    return row['expenses_review']
  elif row['review_status'] == 'sales_with_review':
    return row['expenses_review'] / row['quantity_id']
  else:
    return 0

orders['expenses_review_2'] = orders.apply(expenses_review, axis=1)

# 13 Присоеденим к таблице с заказами затраты на закрепление отзыва
# 13.1 Создадим агрегирующую таблицу
review_fixing = review_fixing.merge(sku, how = 'left', on = 'sku')
review_fixing = review_fixing.groupby(['date_order','article_id']).agg({'expenses_fixing':'sum'}).reset_index()

# 13.2 Присоеденим к таблице заказов данные по затратам на закрепление отзыва
orders = orders.merge(review_fixing, how='outer', on=['date_order','article_id'])

# 13.3 Создадим функцию, которая присвоит статусы закреплениям
def fixing(row):
  if pd.isna(row['order_id']) and pd.isna(row['expenses_fixing']):
    return 'fixing_without_sales'
  elif pd.isna(row['order_id']) and row['expenses_fixing']:
    return 'fixing_without_sales'
  elif pd.isna(row['date_order']) and pd.isna(row['expenses_fixing']):
    return 'empty'
  elif pd.isna(row['order_id']) and row['expenses_fixing'] >= 0:
    return 'empty'
  elif row['quantity'] >= 0 and row['expenses_fixing'] > 0:
    return 'sales_with_fixing'
  elif row['quantity'] >= 0 and pd.isna(row['expenses_fixing']):
    return 'sales_without_fixing'
  else:
    return 'error'

orders['fixing_status'] = orders.apply(fixing, axis=1)

# 13.4 Создадим функцию, которая распределит бонусы по заказам
def expenses_fixing(row):
  if row['fixing_status'] == 'fixing_without_sales':
    return row['expenses_fixing']

  elif row['fixing_status'] == 'sales_without_fixing':
    return row['expenses_fixing']

  elif row['fixing_status'] == 'empty':
    return row['expenses_fixing']

  elif row['fixing_status'] == 'sales_with_fixing':
    return row['expenses_fixing'] / row['quantity_id']

  else:
    return 0

orders['expenses_fixing_2'] = orders.apply(expenses_fixing, axis=1)

# 13.5 Создадим колонку с налогами и чистой прибылью
orders['taxes'] = orders.final_price * 0.06
orders['net_profit'] = orders.gross_profit + orders.taxes - orders.expenses_search - orders.expenses_target_2 - orders.expenses_review_2 - orders.expenses_fixing_2

# 13.6 Создадим функцию, которая определит маржинальность продуктов
def marginality_value(row):
  if row['final_price'] > 0:
    return row['net_profit'] /  row['final_price'] * 100

  elif row['final_price'] < 0:
    return row['net_profit'] /  row['final_price'] * 100

  elif row['fixing_status'] == 0:
    return 0

  else:
    return 0

orders['marginality'] = orders.apply(marginality_value, axis=1)
orders['marginality'] = orders['marginality'].round(2)

# 13.7 Создадим функцию, которая определит ROI каждого продукта
def roi_value(row):
  if row['cost_price_2'] > 0:
    return row['net_profit'] /  row['cost_price_2'] * 100

  elif row['cost_price_2'] < 0:
    return row['net_profit'] /  row['cost_price_2'] * 100

  elif row['cost_price_2'] == 0:
    return 0

  else:
    return 0

orders['roi_value'] = orders.apply(roi_value, axis=1)
orders['roi_value'] = orders['roi_value'].round(2)

orders[['revenue', 'quantity','final_price', 'commission_sale','process_delivery', 'refund_cancel','cost_price', 'cost_price_2','gross_profit', 'expenses_search','expenses_target_2', 'expenses_review_2','expenses_fixing_2']] = orders[['revenue', 'quantity','final_price', 'commission_sale','process_delivery', 'refund_cancel','cost_price', 'cost_price_2','gross_profit', 'expenses_search','expenses_target_2', 'expenses_review_2','expenses_fixing_2']].fillna(0)
print('Результирующая таблица с заказами загружена')
print('Минимальная дата в таблице заказов',orders.date_order.min())
print('Максимальна дата в таблице заказов',orders.date_order.max())
```

```python
# 18.1 Введите начальную и конечную дату, по которым необходим отчет
start_date_articles = '2023-10-20' # начальная дата отчета (формат даты: гггг-мм-дд)
end_date_articles = '2024-03-20' # конечная дата отчета (формат даты: гггг-мм-дд)

# 18.2 Сформируем необходимую таблицу для анализа
orders_articles = orders.copy()
orders_articles = orders_articles.query('date_order >= @start_date_articles and date_order <= @end_date_articles')
```

```python
# 18.5 Рассчитаем кол-во штук по каждой категории за период (quantity)
orders_articles_quantity = orders_articles.copy()
orders_articles_quantity = orders_articles_quantity.groupby('article_id').agg({'quantity':'sum'}).reset_index()

# 18.6 Рассчитаем сумму за продажу или возврат до вычета комиссий и услуг (final_price)
orders_articles_final_price = orders_articles.copy()
orders_articles_final_price = orders_articles_final_price.groupby('article_id').agg({'final_price':'sum'}).reset_index()

# 18.7 Рассчитаем сумму за обработку и доставку (process_delivery)
orders_articles_process_delivery = orders_articles.copy()
orders_articles_process_delivery = orders_articles_process_delivery.groupby('article_id').agg({'process_delivery':'sum'}).reset_index()

# 18.8 Рассчитаем сумму за возвраты и отмены (process_delivery)
orders_articles_refund_cancel = orders_articles.copy()
orders_articles_refund_cancel = orders_articles_refund_cancel.groupby('article_id').agg({'refund_cancel':'sum'}).reset_index()

# 18.9 Рассчитаем сумму итого (revenue)
orders_articles_refund_revenue = orders_articles.copy()
orders_articles_refund_revenue = orders_articles_refund_revenue.groupby('article_id').agg({'revenue':'sum'}).reset_index()

# 18.10 Рассчитаем сумму валовой прибыли (gross_profit)
orders_articles_gross_profit = orders_articles.copy()
orders_articles_gross_profit = orders_articles_gross_profit.groupby('article_id').agg({'gross_profit':'sum'}).reset_index()

# 18.11 Рассчитаем затраты на рекламу (поиск) (expenses_search)
orders_articles_expenses_search = orders_articles.copy()
orders_articles_expenses_search = orders_articles_expenses_search.groupby('article_id').agg({'expenses_search':'sum'}).reset_index()

# 18.12 Рассчитаем затраты на рекламу (таргет) (expenses_search)
orders_articles_expenses_target_2 = orders_articles.copy()
orders_articles_expenses_target_2 = orders_articles_expenses_target_2.groupby('article_id').agg({'expenses_target_2':'sum'}).reset_index()

# 18.13 Рассчитаем затраты на отзывы за баллы (expenses_review_2)
orders_articles_expenses_review_2 = orders_articles.copy()
orders_articles_expenses_review_2 = orders_articles_expenses_review_2.groupby('article_id').agg({'expenses_review_2':'sum'}).reset_index()

# 18.14 Рассчитаем затраты на закрепление отзыва (expenses_review_2)
orders_articles_expenses_fixing_2 = orders_articles.copy()
orders_articles_expenses_fixing_2 = orders_articles_expenses_fixing_2.groupby('article_id').agg({'expenses_fixing_2':'sum'}).reset_index()

# 18.15 Рассчитаем налоги (taxes)
orders_articles_expenses_taxes = orders_articles.copy()
orders_articles_expenses_taxes = orders_articles_expenses_taxes.groupby('article_id').agg({'taxes':'sum'}).reset_index()

# 18.16 Рассчитаем чистую прибыль (net_profit)
orders_articles_net_profit = orders_articles.copy()
orders_articles_net_profit = orders_articles_net_profit.groupby('article_id').agg({'net_profit':'sum'}).reset_index()

# 18.17 Рассчитаем среднюю маржинальность (marginality)
orders_articles_marginality = orders_articles.copy()
orders_articles_marginality = orders_articles_marginality.groupby('article_id').agg({'marginality':'mean'}).reset_index()

# 18.18 Рассчитаем среднее значение ROI (roi_value)
orders_articles_roi_value = orders_articles.copy()
orders_articles_roi_value = orders_articles_roi_value.groupby('article_id').agg({'roi_value':'mean'}).reset_index()

# 18.19 Рассчитаем себестоимость (cost_price_2)
orders_articles_cost_price_2 = orders_articles.copy()
orders_articles_cost_price_2 = orders_articles_cost_price_2.groupby('article_id').agg({'cost_price_2':'sum'}).reset_index()

# 18.20 Соеденим таблицы

orders_articles_all = orders_articles_quantity.merge(orders_articles_final_price, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_process_delivery, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_refund_cancel, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_refund_revenue, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_cost_price_2, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_gross_profit, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_expenses_search, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_expenses_target_2, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_expenses_review_2, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_expenses_fixing_2, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_expenses_taxes, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_net_profit, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_marginality, how='left', on='article_id')
orders_articles_all = orders_articles_all.merge(orders_articles_roi_value, how='left', on='article_id')

orders_articles_all
```

```python
# 19 Рассчитаем показатели магазина по артикулам, по данным за весь период

print('Кол-во артикулов:',orders_articles_all.article_id.count(),"штук")
print('Кол-во штук:',orders_articles_all.quantity.sum(),"штук")
print('За продажу или возврат до вычета комиссий и услуг:',orders_articles_all.final_price.sum(),"рублей")
print('Обработка и доставка (Сборка заказа + Обработка отправления + Магистраль + Последняя миля + Обработка отмененного или невостребованного товара + Обработка невыкупленного товара + Логистика):',orders_articles_all.process_delivery.sum(),"рублей")
print('Возвраты и отмены(Обратная магистраль + Обработка возврата + Обратная логистика):',orders_articles_all.refund_cancel.sum(),"рублей")
print('Итого:',orders_articles_all.revenue.sum().round(2),"рублей")
print('Валовая прибыль:',orders_articles_all.gross_profit.sum(),"рублей")
print('Затраты реклама Поиск:',orders_articles_all.expenses_search.sum().round(2),"рублей")
print('Затраты реклама Таргет:',orders_articles_all.expenses_target_2.sum().round(2),"рублей")
print('Затраты Отзывы за баллы:',orders_articles_all.expenses_review_2.sum().round(2),"рублей")
print('Закрепление отзыва:',orders_articles_all.expenses_fixing_2.sum().round(2),"рублей")
print('Налоги(6%):',orders_articles_all.taxes.sum().round(2),"рублей")
print('Чистая прибыль:',orders_articles_all.net_profit.sum().round(2),"рублей")
print('Маржинальность(%):',((orders_articles_all.net_profit.sum() / orders_articles_all.final_price.sum()) * 100).round(2),"%")
print('ROI(%):',((orders_articles_all.net_profit.sum() / orders_articles_all.cost_price_2.sum()) * 100).round(2),"%")
```
